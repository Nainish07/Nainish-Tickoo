```python
import numpy as np
import ntlib as l
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
colourMap =  plt.cm.gnuplot  # choosing colour  
colorinterpolation = 30   # selecting no. of lines for contour
z =20 # selecting grid length
V_xy ,A = l.Laplace(20,20,20,20,z)    
L = l.jacobi(V_xy,A) # using gauss-jacobi to get the potential values
L = l.matrix(z,z,L) # creating a matrix for the potential values
x_vals = np.linspace(0, 20, z)   
y_vals = np.linspace(0, 20, z)   #poltting the contour map
X, Y = np.meshgrid(x_vals, y_vals)
cp = plt.contourf(X, Y, L, colorinterpolation, cmap=colourMap)
# nt = plt.axes(projection='3d')
# nt.contour3D(X, Y, L)
plt.colorbar(cp)      
plt.title("Contour of Potential for a charged box")    

plt.show()
```
