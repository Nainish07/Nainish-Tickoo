```python
#questioin 1
# equation of the ellipse 

# a is semi minor axis = 1
# b is semi major axis = 2
a = 1
b = 2

def ellipse(n,x0,a,b):
    x = []
    y = []
    ctr = 0
    ec = (((b**2)-(a**2))/a)**(0.5)      #ec is the eccentricity of the ellipse
    constant = (1+(ec))**2+(1-(ec))**2           #any coordinate that lies on the ellipes has sum of
    for i in range(0,n):                         # distances drawn from the foci to be constant
        x.append(2*(l.My_Random(l.My_Random(x0))/32768)) #for x coordinate
        y.append(l.My_Random(x0)/32768) # for y coordinate
        x0 = l.My_Random(x0)
        d = (x[i])**2 + y[i]**2  
        if d < (constant**(0.5)): #check for distance to be less than constant
            ctr = ctr + 1
    ctr = 2*ctr 
    plt.scatter((x),y)
    plt.show()  
    return(4*ctr/n)
   
    

area = ellipse(1000,10,a,b)  #seed 

print(area,"is the calculated value of area of ellipse")
# To check the limit of error
error = (area/(2*math.pi)-1)*100
print(error,"% is the error in the value of the area")    
```
