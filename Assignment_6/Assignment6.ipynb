```python
import ntlib as l
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Question 1
def dxdt(v,t):
    return(v)  #define the function for both single and double derivative

def dvdt(x,v,t):
    return( math.cos(1.2*x)-0.1*v - x)

X1, V1 , T1= l.shm(0,2,-1, 0.5, 20,dvdt,dxdt)  # basically using rk4 to solve
                                               # but name is changed to shm for simplicity
plt.axhline(y = 0, color = 'r', linestyle = '-')

plt.plot(T1, X1, color = 'b', linestyle = '--')
plt.scatter(T1, X1, color = 'g')          #plotting the curve
plt.show()
```

```python
#Question 2
def dxdt(v,t):
    return(v)
                        #define the function for both single and double derivative 
def dvdt(x,v,t):
    return( 2*x)

x1 = 200

vl = 60   # definig vl
X1,V1,T1 = l.rk4_li(0, 40, vl, 0.05, 1, 200,dvdt,dxdt)
         # defining vh
vh = 80
X2,V2,T2 = l.rk4_li(0, 40, vh, 0.05, 1, 200,dvdt,dxdt)

v = l.interpolation_l(vl,vh,X1,V1,T1,X2,V2,T2,x1)
print(v)                                             # using lagrange interpolation to get correct"v"

X3,V3,T3 = l.rk4_li(0, 40, v, 0.05, 1, 200,dvdt,dxdt)

plt.plot(T1, X1, color = 'b')
plt.scatter(T1, X1, color = 'b')

plt.plot(T2, X2, color = 'r')
plt.scatter(T2, X2, color = 'r')

plt.plot(T3, X3, color = 'g')
plt.scatter(T3, X3, color = 'g')

print(T3[11],"is the point at which temperature is 100")
            
plt.show()         # plotting the curves
```
