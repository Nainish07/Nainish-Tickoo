```python
import ntlib as l
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Question 1
g = 0.2
omega = 1.2

def dydt(t, y, v):
    return v
def d2ydt2(t, y, v):
    return (2*math.cos(omega*t)-(g*v)-(2*y))/2


t0_1, x0_1, z0_1 = l.sho_2(d2ydt2, dydt, 0, 2, -1, 50, 0.5)
plt.plot(t0_1,x0_1)
plt.scatter(t0_1,x0_1)
plt.show()
```

```python
#Question 2
def dxdt(v,t):
    return(v)
                        #define the function for both single and double derivative 
def dvdt(x,v,t):
    f = (x-20)
    return(f)

x1 = 200

vl = 60   # definig vl
X1,V1,T1 = l.rk4_li(0, 40, vl, 0.05, 1, 200,dvdt,dxdt)
         # defining vh
vh = 180
X2,V2,T2 = l.rk4_li(0, 40, vh, 0.05, 1, 200,dvdt,dxdt)

v = l.interpolation_l(vl,vh,X1,V1,T1,X2,V2,T2,x1)
print(v)                                             # using lagrange interpolation to get correct"v"

X3,V3,T3 = l.rk4_li(0, 40, v, 0.05, 1, 200,dvdt,dxdt)

plt.plot(T1, X1, color = 'b')
plt.scatter(T1, X1, color = 'b')

plt.plot(T2, X2, color = 'r')
plt.scatter(T2, X2, color = 'r')

plt.plot(T3, X3, color = 'g')
plt.scatter(T3, X3, color = 'g')

plt.axhline(y = 200, color = 'y', linestyle = '-')
plt.axhline(y = 100, color = 'y', linestyle = '-')
plt.axhline(y = 0, color = 'y', linestyle = '-')
print(T3[9],"is the point at which temperature is 100")
            
plt.show()         # plotting the curves
```
