```python
#question 1 using LU- decomposition

with open('q1_(a).txt', 'r') as n:
    z = [[float(num) for num in line.split(',')] for line in n]
print(z)

print("")



n2 = open('q1_(b).txt', 'r')
b= n2.read().split(',')
for i in range(0,len(b)):
    b[i]=float(b[i])
print(b)
print("")

L,U = l.LU_decomposition(z)
print("")
print(L)
print("")

print(U)
print("")

Y=l.forward(L,b)
# print(Y)
X=l.backward(U,Y)
# print("")
print(X)
```

```python
#question 2 using choleski
with open('q2_(a).txt', 'r') as n:
    a = [[float(num) for num in line.split(',')] for line in n]
print(a)
print("")

n2 = open('q2_(b).txt', 'r') 
b= n2.read().split(',')
for i in range(0,len(b)):
    b[i]=float(b[i])
print(b)
print("")

L = l.cholesky(a)
U = l.Transpose_sq(a)
l.prnt_matrix(L)
l.prnt_matrix(U)
Y=l.forward(L,b)
X=l.backward(U,Y)
print(X)
```

```python
#question 3 using LU- decomposition

with open('q3_(a).txt', 'r') as n:
    c = [[float(num) for num in line.split(',')] for line in n]
print(c)

print("")

n2 = open('q3_(b).txt', 'r')
b= n2.read().split(',')
for i in range(0,len(b)):
    b[i]=float(b[i])
print(b)
print("")


L,U = l.LU_decomposition(c)
Y=l.forward(L,b)
X=l.backward(U,Y)
print(X)
```
